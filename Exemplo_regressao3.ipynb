```python
#separa os dados em X e Y, sendo X as features e y a variável target
X = df.drop('thalachh', axis=1)
y = df['thalachh']

#Método de Estatística para Cálculo da Regressão
X_sm = sm.add_constant(X)

results = sm.OLS(y, X_sm).fit() #esse método está sendo executado no conjunto inteiro de dados, ou seja, nos 303 registros
# mostrando as estatísticas do modelo
print(results.summary())
```

```python
heart_df = pd.DataFrame(df)
colunas = df.columns.drop('thalachh')


#simulando o modelo de regressão linear usando a biblioteca sklearn 
lr = LinearRegression()
lr.fit(X, y) #realiza o treinamento do modelo de regressão linear com todos os dados


coef_df = pd.DataFrame(index=np.append(['Intercept'], colunas)) #mostra uma tabela
coef_df['Regressão Linear'] = np.append([lr.intercept_], lr.coef_) 
coef_df.T
```

```python
#análise de métricas do modelo de regressão linear

mse_ls = mean_squared_error(y, lr.predict(X)) #calcular o MSE com o valor predito de x
r2_ls = r2_score(y, lr.predict(X)) #calcular o r^2 do valor predito de x

result_df = pd.DataFrame(index=['MSE', 'R2'])
result_df['Regressão Linear'] = [mse_ls, r2_ls]
result_df
```

```python
from sklearn.linear_model import Lasso
#Método lasso com parâmetro de  alpha=0,05
alpha = 0.01

las = Lasso(alpha=alpha)
las.fit(X, y)

mse_lasso = mean_squared_error(y, las.predict(X))
r2_lasso = r2_score(y, las.predict(X))

result_df['LASSO'] = [mse_lasso, r2_lasso]
result_df
```

```python
#define o parãmetro para regressão de 2o. ordem
col2 = ['{0}\u00b2'.format(c) for c in list(X.columns)]
df2 = X.copy()**2
df2.columns = col2

x2 = pd.concat([X, df2], axis=1)

lr = LinearRegression()
lr.fit(x2, y)

coef_df_2 = pd.DataFrame(index=np.append(['Intercept'], x2.columns))
coef_df_2['Regressão Linear X\u00b2'] = np.append([lr.intercept_], lr.coef_) 
coef_df_2
```

```python
#regressão linear de 2a. ordem
mse_ls = mean_squared_error(y, lr.predict(x2))
r2_ls = r2_score(y, lr.predict(x2))

result_df['Regressão Linear X\u00b2'] = [mse_ls, r2_ls]
result_df
```

```python
alpha = 0.001

las = Lasso(alpha=alpha)
las.fit(x2, y)

coef_df_2['LASSO X\u00b2'] = np.append([las.intercept_], las.coef_) 
coef_df_2
```

```python
#Regressão linear com lasso de 2a. ordem

mse_lasso = mean_squared_error(y, las.predict(x2))
r2_lasso = r2_score(y, las.predict(x2))

result_df['LASSO X\u00b2'] = [mse_lasso, r2_lasso]
result_df
```
